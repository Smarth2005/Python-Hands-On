### 17. Iterables vs Iterators vs Generators

##### Iterable:
An iterable is any object in Python that you can loop through — one item at a time, i.e. meant for iteration. E.g: list, tuple, string etc.

````python
## List is iterable
lst = [1,2,3,4,5]
for i in lst:
    print(i)
````

##### Iterator:
An iterator is an object that fetches one value at a time from an iterable using the next() function. It keeps track of where you are in an iterable (like a list) so that next() knows what to return next.

````python
## iterator
lst1 = iter(lst)
for i in lst1:
    print(i)
````

````python
## another way of retrieving elements through an iterator
lst2 = iter(lst)
type(lst2)
````

````python
next(lst2)
````

````python
print(next(lst2))
print(next(lst2))
print(next(lst2))
print(next(lst2))
print(next(lst2))
````

- Iterators don’t store all values at once — they yield one value at a time.
- Fetching one value at a time makes iterators memory-efficient, especially when working with large datasets.
- When we run a for loop, Python automatically creates an iterator from the iterable and keeps calling `next()` in the background until the iterable is exhausted.

````python
lst3 = iter([12,23,34,45,56,67,78,89,90])
while True:
    try:
        print(next(lst3))  # manually fetch next item
    except StopIteration:
        print("The iterator is empty")
        break
````

##### Generator:
A generator is a special type of iterator that lets you generate values on the fly (meaning it creates each value only when it’s needed, instead of preparing and storing all values in advance.)using the `yield` keyword instead of `return`.
It doesn’t store all values in memory — it yields one value at a time, making it highly memory-efficient.

````python
def square(n):
    for i in range(n):
        return i**2

square(3)
````

Using `return` inside the loop immediately exits the function after returning the first value. So it only returns 0**2 = 0 and stops there.



````python
## Generator
def square(n):
    for i in range(n):
        yield i**2

square(3)
````

Using `yield` instead of `return` turns the function into a generator. It doesn’t run immediately — it returns a generator object.<br>
Values will be produced one-by-one when you iterate over it.

````python
for i in square(3):
    print(i)
````

This is how we consume the generator.

````python
a = square(4)
a
````

````python
next(a)
````

````python
while True:
    try:
        print(next(a))
    except StopIteration:
        break
````

This is a manual way to loop through a generator. It uses `next()` repeatedly until all values are consumed. Once exhausted, a `StopIteration` is raised, and we break out of the loop.



##### Differences between an iterator and a generator
1. Both iterators and generators support lazy evaluation (one value at a time).
2. To create iterator, we use `iter()` and to create generator, we use function along with `yield` keyword.
3. Generator uses the yield keyword: In generators, Python automatically saves the function’s local state (like variables and position) between yield calls. In iterators, we must manage the internal state manually (e.g., using variables in a class).
4. Generator in Python helps us write fast and compact code. Iterators require more boilerplate code.
5. Generators are more memory-efficient, as they yield one value at a time without storing all values in memory.Iterators can also be memory-efficient but require manual implementation.
6. Iterators can be reused by calling iter( ) again. Generators are single-use; once exhausted, we must recreate them.
