##### Function Overloading Allowed?

##### Returning multiple values (tuple)

````python
lst = [1,2,3,4,5,6,7,8]

def even_oddSum(lst):
    even_sum = 0
    odd_sum = 0
    for i in lst:
        if i%2==0:
            even_sum += i
        else:
            odd_sum += i
    return even_sum, odd_sum
````

````python
even_oddSum(lst)
````

##### Positional and Keyword Arguments

IMPORTANT NOTE:<br>
✔️ Arguments are passed to a function (during the call)<br>
✔️ Parameters are received by the function (in the definition)

```python
def gift_box(item):                      # 'item' is a parameter
    print(f"Thank you for the {item}!")

gift_box("chocolates")                   # "chocolates" is an argument
```

In Python, when calling a function, arguments can be passed in two ways: Positional Arguments and Keyword Arguments.

````python
def gift_box(item, sender):
    print(f"You received {item} from {sender}!")

gift_box("mangoes","Aarav")
````

Here, "mangoes" goes to item and "Aarav" goes to sender because of their position. Hence, these are "positional arguments".
These arguments are matched by their position. If I reverse the order, we get an insensible output.

````python
gift_box("Aarav","mangoes")
````

If I now specify the parameter names in the function call, then these arguments would be called "Keyword Arguments".

````python
gift_box(item = "mangoes", sender = "Aarav") # keyword arguments
gift_box(sender = "Aarav", item = "mangoes")
````

Keyword arguments are passed by explicitly naming the parameter in the function call. Order doesn’t matter, as long as the correct parameter names are used. In the second line of output in the above cell, even if the order of arguments is reversed, Python still knows which value goes where because we’re naming the parameters explicitly.



##### Mixing Positional and Keyword Arguments

````python
gift_box("mangoes", sender = "Aarav")   
````

````python
gift_box(sender= "Aarav", "mangoes")   
````

````python
gift_box(item= "mangoes", "Aarav")
````

KEY POINT: Positional arguments come first; then keyword arguments follow. We cannot place a positional argument after a keyword argument.

##### Using *args and **kwargs
`*args`&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;: Captures all extra positional arguments into a tuple.<br>
`**kwargs`: Captures all extra keyword arguments into a dictionary.

````python
def introduce(*args,**kwargs):
    print(args)   # positional arguments
    print(kwargs) # keyword arguments

introduce("Smarth","Kaushal",age=19,dob=2005)
````

````python
lst = ['Smarth', 'Kaushal']
dict_args = {'age': 19, 'dob': 2005}

introduce(lst, dict_args)
````

`lst` is passed as a single positional argument → So it becomes the first element in the `args` tuple. So, `args = (['Smarth', 'Kaushal'], ...)`

`dict_args` is passed as-is (without **) → So it becomes the second element in the same `args` tuple.
Since we didn’t unpack the dictionary using **, nothing goes into `kwargs`.

````python
introduce(*lst, **dict_args)
````

`*lst` → unpacks the list into separate positional arguments → `args = ('Smarth', 'Kaushal')`<br>
`**dict_args` → unpacks the dictionary into keyword arguments → `kwargs = {'age': 19, 'dob': 2005}`<br>

Unpacking means breaking a collection (like a list, tuple, or dictionary) into individual elements so they can be passed as separate arguments to a function.

#### Lambda Functions
A lambda function in Python is a small, anonymous function defined without using the `def` keyword. It is used when you need a short function for one-time or inline use.

- It can take any number of arguments, but contains only a single expression.
- The expression is automatically returned, so no return statement is needed.
- Lambda functions are often used with functions like `map()`, `filter()`, and `sorted()`.

Syntax:
```python
lambda arguments: expression
```

````python
def addition(a,b):
    return a+b
````

````python
addition(4,5)
````

````python
my_sum = lambda a,b: a+b
````

````python
my_sum(12,14)
````

````python
def isEven(num):
    if num%2==0:
        return True

isEven(25)
````

````python
isEven(24)
````

````python
even1 = lambda a: a%2==0
````

````python
even1(24)
````

````python
even1(25)
````

#### Map Functions
The `map()` function in Python applies a given function to each item in an iterable (like a list) and returns a map object (which can be converted to a list). It takes two arguments: a function and an iterable, and returns a new iterable with the function applied to each element.

Syntax:
```python
map(function, iterable)
```

````python
def evenOdd(num):
    if num%2==0:
        return "{} is Even".format(num)
    else:
        return "{} is Odd".format(num)
````

````python
evenOdd(24)
````

````python
lst = [1,2,3,4,5,6,7,8,9,12,23,34,45,56,67,78,89,90]
````

````python
map(evenOdd, lst) 
````

##### lazy loading technique:
When you use map(), it doesn’t compute the results immediately.
Instead, it returns a map object (an iterator), which calculates each value only when you iterate over it.

````python
list(map(evenOdd, lst))
````

````python
def square(n):
    print(f"Calculating square of {n}")
    return n * n

nums = [1, 2, 3, 4, 5]

result = map(square, nums)  # returns a lazy iterator
print(result)
````

Nothing is printed yet — because square() is not called until we force iteration.

````python
print(list(result))
````

#### Filter Functions
The `filter()` function is used to filter elements of an iterable (like a list) based on a function that returns True or False.
It returns an iterator — so it’s also a lazy function, like map( ).

Syntax:
```python
filter(function, iterable)
```

````python
def isEven(num):
    if num%2==0:
        return True
````

````python
lst=[12,23,34,45,56,67,78,89,90]
````

````python
filter(isEven, lst)
````

````python
list(filter(isEven, lst))
````

````python
list(filter(lambda x: x%2==0, lst))
````

````python
list(map(lambda x: x%2==0, lst))
````

````python
fruits = ['apple', 'banana', 'kiwi', 'avocado', 'cherry', 'apricot']

# Filter fruits that start with 'a'
result = filter(lambda fruit: fruit.lower().startswith('a'), fruits)

print(result)              # Just an iterator object
print(list(result))        # Forces evaluation
````

#### Reduce Functions
`reduce()` applies a function cumulatively to the items of an iterable, reducing it to a single value.
Syntax:
```python
from functools import reduce
reduce(function, iterable[, initializer])
```

````python
from functools import reduce
````

````python
numbers = [1,2,3,4,5,6]
my_sum = reduce(lambda x,y: x+y, numbers)
print(my_sum)
````

````python
## Another example
fruits = ['apple', 'banana', 'cherry']

sentence = reduce(lambda x, y: x + ', ' + y, fruits)
print("I like " + sentence + ".")
````

````python
## demo with initializer
numbers = [1, 2, 3]

result = reduce(lambda x, y: x + y, numbers, 10) # 10 is initializer, that means specified operation will be start from number 10
print(result) # 16
````
